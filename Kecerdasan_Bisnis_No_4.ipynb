```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
df=pd.read_excel('/Data Bank.xlsx', 'Data')
df.head()
```

```python
df[['ID', 'Age', 'Income']][(df['Personal Loan']==1) & (df['CCAvg']> 0.0)]
```

```python
df.loc[(df['Personal Loan']==1) & (df['CCAvg']>0.0) & (df['CreditCard']==1)]
```

```python
test = df.loc[(df['Personal Loan']==1) & (df['CCAvg']>0.0) & (df['CreditCard']==1) & (df['Securities Account']==1),['ID', 'Age', 'Income', 'CCAvg','Personal Loan','CreditCard','Securities Account']]
print('Berikut adalah data top 10 nasabah dengan rata-rata CCAvg tertinggi serta memenuhi syarat')
test.nlargest(n=10, columns=['CCAvg'])
```

```python
plt.figure(figsize=(8,6))
for i in range(0, test.shape[0]):
  plt.scatter(df["ID"][i], df["CCAvg"][i], color='Blue')

plt.title('Analysis')
plt.xlabel('ID')
plt.ylabel('CCAvg')
plt.show()
```

```python
test.nlargest(n=10, columns=['CCAvg']).plot.bar(y='CCAvg',x='ID', rot=0)
plt.title('Top 10 of CCAvg', size=16)
plt.ylabel('CCAvg', size=14)
plt.xlabel('\n ID', size=14)
plt.show()
```
