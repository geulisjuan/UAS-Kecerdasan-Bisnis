```python
corpus = [
    'The 2022 FIFA World Cup just became a coronation',
    'Football almost never delivers perfect narratives, so when it happens its all the more impactful',
    'Industry 4.0 opens new challenges for companies',
    'Lionel Messi, at 35, has finally won the World Cup with Argentina, leading his national side to an epic victory over the 2018 champions France',
    'And we can, more comfortably now, crown him the finest mens footballer of all time',
    'It was only fitting that Messis heir apparent, Kylian Mbappe, surged towards the end of this game, scoring three goals in a losing side',
    'Messi had been the best player in the tournament but was fading towards the end of the final',
    'It seems destiny got in the way to deliver him the ultimate prize',
    'The only asterisk that hovered over his legacy was a failure to win the biggest prize in the sport — the World Cup Trophy',
    'That failure is now a gleaming success story',
    'By five years into his professional career, it was obvious Messi was dancing alongside the gods',
    'As the best player at perhaps the best club side the world has seen in Barcelona, he racked up goals, assists, trophies and personal plaudits',
    'To list his most astonishing statistics, he has nearly 800 career goals, 41 trophies and an unmatched seven Ballon dOrs',
    'For a player who has never shown excessive flair, he has been irresistible to watch',
    'Never once in his career has he performed an unnecessary step-over or backheel, yet he is an endless highlights reel of dazzling dribbles',
    'Despite being unrelentingly effective, he s never been dull to watch. Quite the opposite of a robotic winning machine',
    'he s the player most likely to make you gasp in disbelief, or laugh out loud',
    'While Maradona was often likened to a god, and Pelé to a king, Messi sometimes makes us think he is an alien, something otherworldly',
    'Like he s playing a different game to everyone else',
    'There is a strong case that he didnt need to win the World Cup to be regarded as the greatest'
]

corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
vectorizer_X = vectorizer.fit_transform(corpus).todense()
vectorizer_X
```

```python
vectorizer.get_feature_names()
```

```python
from sklearn.metrics.pairwise import euclidean_distances

for i in range(len(vectorizer_X)):
  for j in range(i, len (vectorizer_X)):
    if i == j:
      continue
    jarak = euclidean_distances(vectorizer_X[i], vectorizer_X[j])
    print(f'Jarak Dokumen {i+1} dan {j+1}: {jarak}')
     
```
